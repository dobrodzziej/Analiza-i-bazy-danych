```python
import pandas as pd
import math
```

```python
# This command file produces organised database of tuberculosis cases. Data origins from
# "https://github.com/KAIR-ISZ/public_lectures/tree/master/Analiza%20i%20Bazy%20Danych%202021/Lab%202"
```

```python
df = pd.read_csv (r'C:\Studia\Analiza-i-bazy-danych\Lab2\Original_Data\tb2.csv')
df = pd.melt(df, id_vars=['iso2', 'year', 'new_sp'], value_vars=list(df.columns)[3:],
             var_name='column', value_name='cases')
df = df[pd.isnull(df['cases']) != True]
df = df[pd.isnull(df['new_sp']) != True]
df['cases'] = df['cases'].astype(int)
df['new_sp'] = df['new_sp'].astype(int)
df['sex'] = df['column'].str[7]
df['age'] = df['column'].str[8:].map({
    '04': '0-4',
    '514': '5-14',
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+',
    'u': 'unknown'
})
df.rename(columns={'iso2': 'country', 'new_sp': 'total annual cases'}, inplace=True)
df = df[['country', 'year', 'sex', 'age', 'cases', 'total annual cases']]
df = df.sort_values(["country", "year", "sex", "age"], ascending=True)
df.to_csv(r'C:\Studia\Analiza-i-bazy-danych\Lab2\Original_Data\tb-tidy-full.csv', index=False)
df.head(25)
```
